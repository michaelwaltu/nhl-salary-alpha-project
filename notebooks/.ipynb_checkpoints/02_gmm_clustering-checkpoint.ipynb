```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
# Load the data
df = pd.read_csv('../data/processed/modeling_ready_stats.csv')
```

```python
# Sanity check 
print("Final Data Check:")
print(df.shape)
print("Duplicate column names:", df.columns.duplicated().sum())
print("Duplicate column data:", df.T.duplicated().sum())
```

```python
# Quick look at position distribution
print(df['Pos'].value_counts())
print()
print(df['position'].value_counts())  
```

```python
# Split into forwards and defense (goalies are excluded)
FORWARD_POS  = ['C', 'LW', 'RW', 'F', 'C/LW', 'LW/RW', 'C/RW', 'C/W', 'W', 'RW/C', 'LW/C'] 
DEFENSE_POS  = ['D']

fwd = df[df['Pos'].isin(FORWARD_POS)].copy()
dfd = df[df['Pos'].isin(DEFENSE_POS)].copy()

print(f'Forwards : {len(fwd):,} player-seasons')
print(f'Defenders: {len(dfd):,} player-seasons')
```

```python
# ── Create derived columns on the FULL df first ────────────────────────────────
df['FO_total'] = df['FOW'] + df['FOL']

per_game_cols = [
    'G', 'A', 'PTS', 'EVG', 'PPG',
    'I_F_xGoals', 'I_F_flurryScoreVenueAdjustedxGoals',
    'I_F_primaryAssists', 'I_F_secondaryAssists',
    'I_F_shotsOnGoal', 'I_F_missedShots', 'I_F_blockedShotAttempts',
    'I_F_highDangerShots', 'I_F_mediumDangerShots', 'I_F_lowDangerShots',
    'I_F_highDangerxGoals', 'I_F_xRebounds',
    'I_F_hits', 'I_F_takeaways', 'I_F_giveaways', 'I_F_dZoneGiveaways',
    'shotsBlockedByPlayer', 'PIM', 'FO_total'
]

for col in per_game_cols:
    df[f'{col}_pg'] = df[col] / df['GP']

# ── Re-split AFTER creating new columns ───────────────────────────────────────
FORWARD_POS = ['C', 'LW', 'RW', 'F', 'C/LW', 'LW/RW', 'C/RW', 'C/W', 'W', 'RW/C', 'LW/C']
DEFENSE_POS = ['D']

fwd = df[df['Pos'].isin(FORWARD_POS)].copy()
dfd = df[df['Pos'].isin(DEFENSE_POS)].copy()

# ── Define feature lists ───────────────────────────────────────────────────────
SHARED_FEATURES = [
    'Age',
    'TOI/60',
    'oZS%', 'dZS%',
    'onIce_xGoalsPercentage', 'onIce_corsiPercentage', 'onIce_fenwickPercentage',
    'G_pg', 'A_pg', 'PTS_pg', 'EVG_pg', 'PPG_pg',
    'I_F_xGoals_pg', 'I_F_flurryScoreVenueAdjustedxGoals_pg',
    'I_F_primaryAssists_pg', 'I_F_secondaryAssists_pg',
    'I_F_shotsOnGoal_pg', 'I_F_missedShots_pg', 'I_F_blockedShotAttempts_pg',
    'I_F_highDangerShots_pg', 'I_F_mediumDangerShots_pg', 'I_F_lowDangerShots_pg',
    'I_F_highDangerxGoals_pg', 'I_F_xRebounds_pg',
    'I_F_hits_pg', 'I_F_takeaways_pg', 'I_F_giveaways_pg', 'I_F_dZoneGiveaways_pg',
    'shotsBlockedByPlayer_pg', 'PIM_pg',
]

FWD_FEATURES = SHARED_FEATURES + ['FO%', 'FO_total_pg']
DEF_FEATURES = SHARED_FEATURES

# ── GP filter ─────────────────────────────────────────────────────────────────
MIN_GP = 20
fwd_q = fwd[fwd['GP'] >= MIN_GP].copy()
dfd_q = dfd[dfd['GP'] >= MIN_GP].copy()

print(f'Forwards  before filter: {len(fwd):,}  →  after GP >= {MIN_GP}: {len(fwd_q):,}')
print(f'Defenders before filter: {len(dfd):,}  →  after GP >= {MIN_GP}: {len(dfd_q):,}')

missing_fwd = [f for f in FWD_FEATURES if f not in fwd_q.columns]
missing_def = [f for f in DEF_FEATURES if f not in dfd_q.columns]
print(f'\nMissing from FWD_FEATURES: {missing_fwd}')
print(f'Missing from DEF_FEATURES: {missing_def}')
```

```python
# ── Convert TOI/60 from 'MM:SS' string to decimal minutes ─────────────────────
def toi_to_float(toi_str):
    try:
        parts = str(toi_str).split(':')
        return int(parts[0]) + int(parts[1]) / 60
    except:
        return np.nan

df['TOI/60'] = df['TOI/60'].apply(toi_to_float)

# Re-split since we modified df
fwd = df[df['Pos'].isin(FORWARD_POS)].copy()
dfd = df[df['Pos'].isin(DEFENSE_POS)].copy()
fwd_q = fwd[fwd['GP'] >= MIN_GP].copy()
dfd_q = dfd[dfd['GP'] >= MIN_GP].copy()

# Verify
print(df['TOI/60'].dtype)
print(df['TOI/60'].describe().round(2))
```

```python
# See how many numeric columns we actually have after excluding admin cols
EXCLUDE_COLS = [
    'Rk', 'Player', 'Age', 'Team', 'Pos', 'Season', 'playerId', 
    'position', 'situation', 'TOI(EV)', 'TOI', 'ATOI',
    'Cap_Hit', 'Cap_Pct'
]

numeric_cols = [
    c for c in fwd_q.columns 
    if c not in EXCLUDE_COLS 
    and fwd_q[c].dtype in ['float64', 'int64']
]

print(f'Total numeric columns available for full-feature PCA: {len(numeric_cols)}')
print(numeric_cols)
```

```python
# ── Scale features ─────────────────────────────────────────────────────────────
scaler_fwd = StandardScaler()
scaler_def = StandardScaler()

X_fwd = scaler_fwd.fit_transform(fwd_q[FWD_FEATURES])
X_def = scaler_def.fit_transform(dfd_q[DEF_FEATURES])

print(f'X_fwd shape: {X_fwd.shape}')
print(f'X_def shape: {X_def.shape}')
```

```python
# ── BIC / AIC Sweep ────────────────────────────────────────────────────────────
K_RANGE = range(2, 15)

def sweep_gmm(X, label):
    results = {}
    print(f'\n{label}')
    print(f'{"k":>4}  {"BIC":>12}  {"AIC":>12}')
    print('-' * 32)
    for k in K_RANGE:
        gmm = GaussianMixture(n_components=k, covariance_type='full', n_init=10, random_state=42)
        gmm.fit(X)
        results[k] = {'bic': gmm.bic(X), 'aic': gmm.aic(X), 'model': gmm}
        print(f'{k:>4}  {gmm.bic(X):>12,.1f}  {gmm.aic(X):>12,.1f}')
    return results

fwd_sweep = sweep_gmm(X_fwd, '=== FORWARDS ===')
def_sweep = sweep_gmm(X_def, '=== DEFENSEMEN ===')

# ── Plot BIC / AIC curves ──────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, sweep, title in zip(axes, [fwd_sweep, def_sweep], ['Forwards', 'Defensemen']):
    ks   = list(sweep.keys())
    bics = [sweep[k]['bic'] for k in ks]
    aics = [sweep[k]['aic'] for k in ks]
    
    ax.plot(ks, bics, 'o-', color='steelblue',  linewidth=2, label='BIC')
    ax.plot(ks, aics, 's--', color='darkorange', linewidth=2, label='AIC')
    
    best_bic_k = ks[np.argmin(bics)]
    best_aic_k = ks[np.argmin(aics)]
    ax.axvline(best_bic_k, color='steelblue',  linestyle=':', alpha=0.7, label=f'BIC min → k={best_bic_k}')
    ax.axvline(best_aic_k, color='darkorange', linestyle=':', alpha=0.7, label=f'AIC min → k={best_aic_k}')
    
    ax.set_title(f'{title} — BIC & AIC by k', fontsize=13, fontweight='bold')
    ax.set_xlabel('Number of clusters (k)')
    ax.set_ylabel('Score (lower = better)')
    ax.set_xticks(ks)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# Columns to exclude from full-feature PCA
# Raw counts we have per-game versions of, plus admin/identity cols
PCA_EXCLUDE = [
    # Admin
    'GP', 'games_played', 'icetime', 'shifts', 'iceTimeRank', 'timeOnBench',
    # Raw counts superseded by _pg versions
    'G', 'A', 'PTS', 'EVG', 'PPG', 'SHG', 'GWG',
    'EV', 'PP', 'SH', 'SOG', 'TSA', 'SAtt.',
    'FOW', 'FOL', 'FO_total',
    'BLK', 'HIT', 'TK', 'GV', 'TAKE', 'GIVE', 'PIM',
    'I_F_xGoals', 'I_F_primaryAssists', 'I_F_secondaryAssists',
    'I_F_shotsOnGoal', 'I_F_missedShots', 'I_F_blockedShotAttempts',
    'I_F_highDangerShots', 'I_F_mediumDangerShots', 'I_F_lowDangerShots',
    'I_F_highDangerxGoals', 'I_F_xRebounds', 'I_F_hits',
    'I_F_takeaways', 'I_F_giveaways', 'I_F_dZoneGiveaways',
    'shotsBlockedByPlayer', 'I_F_flurryScoreVenueAdjustedxGoals',
    # Highly redundant adjusted versions
    'I_F_flurryAdjustedxGoals', 'I_F_scoreVenueAdjustedxGoals',
    'I_F_xGoals_with_earned_rebounds', 'I_F_xGoals_with_earned_rebounds_scoreAdjusted',
    'I_F_xGoals_with_earned_rebounds_scoreFlurryAdjusted',
    'I_F_reboundxGoals', 'I_F_xGoalsFromxReboundsOfShots', 'I_F_xGoalsFromActualReboundsOfShots',
    'I_F_scoreAdjustedShotsAttempts', 'I_F_scoreAdjustedUnblockedShotAttempts',
    'I_F_unblockedShotAttempts', 'I_F_shotAttempts', 'I_F_points', 'I_F_goals',
    'I_F_savedShotsOnGoal', 'I_F_savedUnblockedShotAttempts',
    # Raw shift counts (have % versions)
    'I_F_oZoneShiftStarts', 'I_F_dZoneShiftStarts',
    'I_F_neutralZoneShiftStarts', 'I_F_flyShiftStarts',
    'I_F_oZoneShiftEnds', 'I_F_dZoneShiftEnds',
    'I_F_neutralZoneShiftEnds', 'I_F_flyShiftEnds',
    # OffIce (player on bench)
    'OffIce_F_xGoals', 'OffIce_A_xGoals', 'OffIce_F_shotAttempts', 'OffIce_A_shotAttempts',
    'offIce_xGoalsPercentage', 'offIce_corsiPercentage', 'offIce_fenwickPercentage',
    # Cap (not a performance feature)
    'Cap_Hit', 'Cap_Pct',
]

pca_cols = [c for c in numeric_cols if c not in PCA_EXCLUDE]
print(f'Full feature PCA columns: {len(pca_cols)}')
print(pca_cols)
```

```python
# ── Cell 12: Full-feature PCA ─────────────────────────────────────────────────

# Final cleanup of pca_cols — remove anything that slipped through
PCA_ALSO_REMOVE = [
    'OnIce_F_xOnGoal', 'OnIce_F_xGoals', 'OnIce_F_flurryAdjustedxGoals',
    'OnIce_F_scoreVenueAdjustedxGoals', 'OnIce_F_flurryScoreVenueAdjustedxGoals',
    'OnIce_F_shotsOnGoal', 'OnIce_F_missedShots', 'OnIce_F_blockedShotAttempts',
    'OnIce_F_shotAttempts', 'OnIce_F_goals', 'OnIce_F_rebounds', 'OnIce_F_reboundGoals',
    'OnIce_F_lowDangerShots', 'OnIce_F_mediumDangerShots', 'OnIce_F_highDangerShots',
    'OnIce_F_lowDangerxGoals', 'OnIce_F_mediumDangerxGoals', 'OnIce_F_highDangerxGoals',
    'OnIce_F_lowDangerGoals', 'OnIce_F_mediumDangerGoals', 'OnIce_F_highDangerGoals',
    'OnIce_F_scoreAdjustedShotsAttempts', 'OnIce_F_unblockedShotAttempts',
    'OnIce_F_scoreAdjustedUnblockedShotAttempts', 'OnIce_F_xGoalsFromxReboundsOfShots',
    'OnIce_F_xGoalsFromActualReboundsOfShots', 'OnIce_F_reboundxGoals',
    'OnIce_F_xGoals_with_earned_rebounds', 'OnIce_F_xGoals_with_earned_rebounds_scoreAdjusted',
    'OnIce_F_xGoals_with_earned_rebounds_scoreFlurryAdjusted',
    'OnIce_A_xOnGoal', 'OnIce_A_xGoals', 'OnIce_A_flurryAdjustedxGoals',
    'OnIce_A_scoreVenueAdjustedxGoals', 'OnIce_A_flurryScoreVenueAdjustedxGoals',
    'OnIce_A_shotsOnGoal', 'OnIce_A_missedShots', 'OnIce_A_blockedShotAttempts',
    'OnIce_A_shotAttempts', 'OnIce_A_goals', 'OnIce_A_rebounds', 'OnIce_A_reboundGoals',
    'OnIce_A_lowDangerShots', 'OnIce_A_mediumDangerShots', 'OnIce_A_highDangerShots',
    'OnIce_A_lowDangerxGoals', 'OnIce_A_mediumDangerxGoals', 'OnIce_A_highDangerxGoals',
    'OnIce_A_lowDangerGoals', 'OnIce_A_mediumDangerGoals', 'OnIce_A_highDangerGoals',
    'OnIce_A_scoreAdjustedShotsAttempts', 'OnIce_A_unblockedShotAttempts',
    'OnIce_A_scoreAdjustedUnblockedShotAttempts', 'OnIce_A_xGoalsFromxReboundsOfShots',
    'OnIce_A_xGoalsFromActualReboundsOfShots', 'OnIce_A_reboundxGoals',
    'OnIce_A_xGoals_with_earned_rebounds', 'OnIce_A_xGoals_with_earned_rebounds_scoreAdjusted',
    'OnIce_A_xGoals_with_earned_rebounds_scoreFlurryAdjusted',
    'PDO', 'gameScore', 'E+/-', 'xGoalsForAfterShifts',
]

pca_cols = [c for c in pca_cols if c not in PCA_ALSO_REMOVE]
print(f'Final PCA column count: {len(pca_cols)}')

# ── Scale for PCA ──────────────────────────────────────────────────────────────
scaler_pca_fwd = StandardScaler()
scaler_pca_def = StandardScaler()

X_pca_fwd_raw = scaler_pca_fwd.fit_transform(fwd_q[pca_cols])
X_pca_def_raw = scaler_pca_def.fit_transform(dfd_q[pca_cols])

# ── Fit PCA — enough components to explain 90% variance ───────────────────────
pca_fwd = PCA(n_components=0.90, random_state=42)
pca_def = PCA(n_components=0.90, random_state=42)

X_pca_fwd = pca_fwd.fit_transform(X_pca_fwd_raw)
X_pca_def = pca_def.fit_transform(X_pca_def_raw)

print(f'\nForwards : {X_pca_fwd_raw.shape[1]} features → {X_pca_fwd.shape[1]} PCA components (90% variance)')
print(f'Defenders: {X_pca_def_raw.shape[1]} features → {X_pca_def.shape[1]} PCA components (90% variance)')

# ── Plot cumulative variance explained ────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

for ax, pca, title in zip(axes, [pca_fwd, pca_def], ['Forwards', 'Defenders']):
    cumvar = np.cumsum(pca.explained_variance_ratio_)
    ax.plot(range(1, len(cumvar)+1), cumvar, 'o-', color='steelblue', linewidth=2)
    ax.axhline(0.90, color='red', linestyle='--', alpha=0.7, label='90% threshold')
    ax.set_xlabel('Number of components')
    ax.set_ylabel('Cumulative variance explained')
    ax.set_title(f'{title} — PCA Variance', fontsize=12, fontweight='bold')
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# ── Cell 13: t-SNE on PCA components ─────────────────────────────────────────
tsne = TSNE(n_components=2, perplexity=40, random_state=42, max_iter=1000)
tsne_fwd = tsne.fit_transform(X_pca_fwd)

tsne2 = TSNE(n_components=2, perplexity=40, random_state=42, max_iter=1000)
tsne_def = tsne2.fit_transform(X_pca_def)

fwd_q = fwd_q.copy()
dfd_q = dfd_q.copy()
fwd_q['tsne_x'] = tsne_fwd[:, 0]
fwd_q['tsne_y'] = tsne_fwd[:, 1]
dfd_q['tsne_x'] = tsne_def[:, 0]
dfd_q['tsne_y'] = tsne_def[:, 1]

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for ax, df_plot, title in zip(
    axes,
    [fwd_q, dfd_q],
    ['Forwards — t-SNE (PCA components)', 'Defenders — t-SNE (PCA components)']
):
    ax.scatter(
        df_plot['tsne_x'], df_plot['tsne_y'],
        alpha=0.4, s=12, color='steelblue', edgecolors='none'
    )
    ax.set_title(title, fontsize=12, fontweight='bold')
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```

```python
# ── Cell 13b: t-SNE with lower perplexity to surface local structure ──────────
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

for i, perp in enumerate([15, 30]):
    tsne_test = TSNE(n_components=2, perplexity=perp, random_state=42, max_iter=1000)
    coords_fwd = tsne_test.fit_transform(X_pca_fwd)
    
    tsne_test2 = TSNE(n_components=2, perplexity=perp, random_state=42, max_iter=1000)
    coords_def = tsne_test2.fit_transform(X_pca_def)
    
    axes[i][0].scatter(coords_fwd[:, 0], coords_fwd[:, 1],
                       alpha=0.4, s=12, color='steelblue', edgecolors='none')
    axes[i][0].set_title(f'Forwards — perplexity={perp}', fontweight='bold')
    axes[i][0].set_xticks([]); axes[i][0].set_yticks([])
    
    axes[i][1].scatter(coords_def[:, 0], coords_def[:, 1],
                       alpha=0.4, s=12, color='darkorange', edgecolors='none')
    axes[i][1].set_title(f'Defenders — perplexity={perp}', fontweight='bold')
    axes[i][1].set_xticks([]); axes[i][1].set_yticks([])

plt.tight_layout()
plt.show()
```

```python
# ── Cell 14: BIC/AIC sweep on PCA-reduced data ────────────────────────────────
fwd_pca_sweep = sweep_gmm(X_pca_fwd, '=== FORWARDS (PCA) ===')
def_pca_sweep = sweep_gmm(X_pca_def, '=== DEFENSEMEN (PCA) ===')

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for ax, sweep, title in zip(
    axes,
    [fwd_pca_sweep, def_pca_sweep],
    ['Forwards (PCA)', 'Defensemen (PCA)']
):
    ks   = list(sweep.keys())
    bics = [sweep[k]['bic'] for k in ks]
    aics = [sweep[k]['aic'] for k in ks]

    ax.plot(ks, bics, 'o-',  color='steelblue',  linewidth=2, label='BIC')
    ax.plot(ks, aics, 's--', color='darkorange', linewidth=2, label='AIC')

    best_bic_k = ks[np.argmin(bics)]
    best_aic_k = ks[np.argmin(aics)]
    ax.axvline(best_bic_k, color='steelblue',  linestyle=':', alpha=0.7, label=f'BIC min → k={best_bic_k}')
    ax.axvline(best_aic_k, color='darkorange', linestyle=':', alpha=0.7, label=f'AIC min → k={best_aic_k}')

    ax.set_title(f'{title} — BIC & AIC by k', fontsize=13, fontweight='bold')
    ax.set_xlabel('Number of clusters (k)')
    ax.set_ylabel('Score (lower = better)')
    ax.set_xticks(ks)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
# ── Cell 15: Fit candidate k values and inspect cluster profiles ───────────────

INSPECT_FEATURES = [
    'Age', 'TOI/60', 'oZS%', 'dZS%',
    'G_pg', 'A_pg', 'PTS_pg', 'EVG_pg', 'PPG_pg',
    'I_F_xGoals_pg', 'I_F_primaryAssists_pg',
    'I_F_highDangerShots_pg', 'I_F_shotsOnGoal_pg',
    'I_F_hits_pg', 'shotsBlockedByPlayer_pg',
    'I_F_takeaways_pg', 'I_F_giveaways_pg',
    'onIce_xGoalsPercentage', 'onIce_corsiPercentage',
    'FO%', 'FO_total_pg', 'PIM_pg'
]

def fit_and_profile(X, df_q, k, label, feature_cols=INSPECT_FEATURES, use_pca=True):
    """Fit GMM, print cluster profiles and sample players."""
    gmm = GaussianMixture(n_components=k, covariance_type='full', n_init=10, random_state=42)
    gmm.fit(X)
    
    df_out = df_q.copy()
    df_out['cluster'] = gmm.predict(X)
    df_out['cluster_prob'] = gmm.predict_proba(X).max(axis=1)
    
    print(f'\n{"="*60}')
    print(f'{label}  —  k={k}')
    print(f'{"="*60}')
    print(f'Cluster sizes:\n{df_out["cluster"].value_counts().sort_index().to_string()}')
    
    # Profile: mean of inspect features per cluster, sorted by PTS_pg
    valid_feats = [f for f in feature_cols if f in df_out.columns]
    profile = df_out.groupby('cluster')[valid_feats].mean().round(3)
    if 'PTS_pg' in profile.columns:
        profile = profile.sort_values('PTS_pg', ascending=False)
    print(f'\nCluster profiles (sorted by PTS/game):')
    print(profile.to_string())
    
    # Sample 5 most prototypical players per cluster
    print(f'\nSample players per cluster (highest assignment confidence):')
    sample_cols = ['Player', 'Season', 'Team', 'Age', 'GP', 'G', 'A', 'PTS', 'TOI/60', 'cluster', 'cluster_prob']
    sample_cols = [c for c in sample_cols if c in df_out.columns]
    for cid in sorted(df_out['cluster'].unique()):
        grp = df_out[df_out['cluster'] == cid].nlargest(5, 'cluster_prob')[sample_cols]
        print(f'\n  Cluster {cid}:')
        print(grp.to_string(index=False))
    
    return df_out, gmm


# ── FORWARDS: try k = 4, 6, 8 ─────────────────────────────────────────────────
fwd_results = {}
for k in [4, 6, 8]:
    df_out, gmm = fit_and_profile(X_pca_fwd, fwd_q, k, 'FORWARDS')
    fwd_results[k] = {'df': df_out, 'gmm': gmm}

# ── DEFENDERS: try k = 3, 4, 5 ────────────────────────────────────────────────
def_results = {}
for k in [3, 4, 5]:
    df_out, gmm = fit_and_profile(X_pca_def, dfd_q, k, 'DEFENDERS')
    def_results[k] = {'df': df_out, 'gmm': gmm}
```

```python
# Replace zeroed HR columns with MoneyPuck equivalents before re-creating _pg cols
# I_F_goals = total goals (all situations), I_F_points = total points

df['G'] = df.apply(
    lambda r: r['I_F_goals'] if r['G'] == 0 and r['I_F_goals'] > 0 else r['G'], axis=1
)
df['A'] = df.apply(
    lambda r: r['I_F_points'] - r['I_F_goals'] if r['A'] == 0 and r['I_F_points'] > 0 else r['A'], axis=1
)
df['PTS'] = df.apply(
    lambda r: r['I_F_points'] if r['PTS'] == 0 and r['I_F_points'] > 0 else r['PTS'], axis=1
)

# Also fix EVG and PPG using MoneyPuck splits
# I_F_goals has all situations — we don't have a perfect EV/PP split from MP
# but we can at least make sure G/A/PTS are no longer zeroed
# EVG and PPG we'll leave for now — they'll get partially fixed via G being corrected

# Recompute _pg columns after fix
for col in per_game_cols:
    df[f'{col}_pg'] = df[col] / df['GP']

# Re-split and re-filter
fwd = df[df['Pos'].isin(FORWARD_POS)].copy()
dfd = df[df['Pos'].isin(DEFENSE_POS)].copy()
fwd_q = fwd[fwd['GP'] >= MIN_GP].copy()
dfd_q = dfd[dfd['GP'] >= MIN_GP].copy()

# Verify fix
print("Troy Terry after fix:")
print(df[df['Player'] == 'Troy Terry'][['Player', 'Season', 'GP', 'G', 'A', 'PTS', 'G_pg', 'PTS_pg']].to_string())
print("\nMatthew Tkachuk after fix:")
print(df[df['Player'] == 'Matthew Tkachuk'][['Player', 'Season', 'GP', 'G', 'A', 'PTS', 'G_pg', 'PTS_pg']].to_string())

# Recheck zero count
zero_mask = (fwd_q['GP'] >= 20) & (fwd_q['G'] == 0) & (fwd_q['A'] == 0) & (fwd_q['PTS'] == 0)
print(f'\nRemaining zeroed forward seasons after fix: {zero_mask.sum()}')
```

```python
# Re-run scaling on hand-picked features
scaler_fwd = StandardScaler()
scaler_def = StandardScaler()
X_fwd = scaler_fwd.fit_transform(fwd_q[FWD_FEATURES])
X_def = scaler_def.fit_transform(dfd_q[DEF_FEATURES])

# Re-run PCA scaling
scaler_pca_fwd = StandardScaler()
scaler_pca_def = StandardScaler()
X_pca_fwd_raw = scaler_pca_fwd.fit_transform(fwd_q[pca_cols])
X_pca_def_raw = scaler_pca_def.fit_transform(dfd_q[pca_cols])

pca_fwd = PCA(n_components=0.90, random_state=42)
pca_def = PCA(n_components=0.90, random_state=42)
X_pca_fwd = pca_fwd.fit_transform(X_pca_fwd_raw)
X_pca_def = pca_def.fit_transform(X_pca_def_raw)

print(f'X_fwd shape: {X_fwd.shape}')
print(f'X_def shape: {X_def.shape}')
print(f'X_pca_fwd shape: {X_pca_fwd.shape}')
print(f'X_pca_def shape: {X_pca_def.shape}')
```
